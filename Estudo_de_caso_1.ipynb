# Análise de Vendas

Neste estudo de caso foi analisado as vendas de uma empresa fictícia que é uma revendedora de automóveis de luxo em São Paulo atuando desde 2016.
Como Analista de Dados apresentei um Dashboard com os dados de vendas num período de 2016 à 2019, tendo como fonte de dados um arquivo excel com os dados do sistema de vendas e CRM da empresa.

DashboardDeVendas.png

Foi apresentado o total de vendas por ano juntamente com a média de vendas para que o CEO visualizasse se está acima ou abaixo do esperado. Afim de ajudar na decisão se o CEO continua com a venda do automóvel da marca Jaguar,mostrei o custo de mão de obra por Estado e por Fabricante. Sabendo que o assunto também poderia aparecer na apresentação coloquei o total de vendas geral e uma matriz de vendas.
Meu trabalho foi mostrar todas essas informações para o meu Gerente que ia expor os resultados da equipe comercial para o novo CEO da empresa.
