# Definindo Estrátegias de Vendas

Neste estudo de caso, sou Analista de Dados na empresa PontoMaximo, uma rede de varejo que vende produtos eletrônicos e eletrodomésticos atuando desde 2012 nos quatro Estados da região Sudeste mais os Estados do Paraná e Bahia.

A empresa está montando a estratégia de vendas para o próximo ano e precisa saber qual dos fabricantes dos produtos vendidos, apresenta melhor desempenho de vendas, com o objetivo de descartar os fabricantes cujos produtos possuem poucas vendas e tentar negociar melhores condições para os principais fabricantes, criei uma visualização com essas informações.

Como a empresa gostaria de ter diferentes visões das vendas realizadas nos últimos quatro anos (período de 2012 a 2015), segmentei os relatórios de vendas por diferentes informações e por diferentes ângulos dando assim suporte as estratégias de vendas da empresa para o próximo ano.

Sabendo que haverá diversas reuniões para a definição da estratégia de vendas e os relatórios poderão ser extraídos sob demanda, de acordo com a necessidade dos gestores, criei um modelo de dados que permite a extração de relatórios a qualquer momento e que permite extrair dados por diferentes visões e ângulos.

Minha fonte de dados foi um arquivo Excel com dados coletados do sistema de vendas, CRM e ERP da empresa.
