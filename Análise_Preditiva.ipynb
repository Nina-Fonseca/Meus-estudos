# Análise de Vendas de uma Rede de Supermercados

Para esse projeto utilizei o banco de dados do site Kaggle, para uma análise preditiva de uma rede de supermercados num período de três meses. Meu objetivo é informar a receita bruta, satisfação do cliente e funcionamento dos setores.

inicialmente criei uma visualização geral da rede de supermercados mostrando a quantidade de clientes e produtos, satisfação do cliente que se mostra insatisfatória nessa análise e essa informação é importante para o tomador de decisão pois aqui ele pode ver o que pode ser melhorado aumentando assim futuros lucros. Apresento o faturamento geral e por dia, o tipo de pagamento mostra o que os clientes utilizam mais para pagar suas compras. A análise do sexo dos clientes mostra que tanto homens como mulheres compram igualmente. Com o ticket médio o tomador de decisão recebe a informação de receita assim como também qual produto é mais vendido.

A visualização é interativa e o tomador de decisão recebe também as mesmas informações de cada supermercado na sequência.

Essa análise nos faz entender o que aconteceu e o porque aconteceu e com base nessas informações se entende o que deve ser feito.
