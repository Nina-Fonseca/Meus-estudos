# Dashboard de Vendas, Custo, Margem de Lucro e KPI

Essa análise foi feita através de uma planilha Excel fictícia. Para chegar ao resultado final foi feito um trabalho de modelagem, relacionamento e DAX nos dados fornecidos.

Com planilhas extensas a modelagem de dados ajuda bastante e é importante porque se pode alimentar cada tabela com uma fonte diferente de dados. No relacionamento, para evitar informações equivocadas foi criado uma tabela intermediária e com o DAX foi feita a junção de colunas e a criação de uma coluna calculada.

O primeiro gráfico que mostra a média de vendas é muito usado no mundo dos negócios porque ele mostra a evolução de alguma coisa ao longo do tempo. Os gráficos criados são precisos e objetivos pois não requerem muita interpretação. Para o gráfico de margem de lucro foi usado a coluna calculada feito anteriormente com o DAX. 

O gráfico valor de venda x preço de custo é um gráfico de dispersão amplamente usado em Machine Learning onde mostra o relacionamento entre duas variáveis. No gráfico KPI utilizei novamente a coluna calculada para que pudesse ser definido uma meta de margem de lucro e como se pode ver a meta está quase sendo alcançada, o dia é a medida de tempo usada para avaliar exatamente a evolução do indicador.
