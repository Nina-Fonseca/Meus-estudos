# Analise de Dados de E-commerce com Power BI e Google Analytics 


Qualquer empresa que tenha um portal, um web site, que tenha um serviço de vendas on-line pode ser auxiliada com o Google Analytics, pois no e-commerce o clique de um usuário é valioso.
A empresa em questão comercializa produtos on-line e configurou o Google Analytics para extrair dados sobre o perfil dos visitantes do portal. Com esse Dashboard o Gestor do setor de Marketing tem informações essenciais como:

1. Por onde os clientes acesam o portal, se é por busca ôrganica ou paga, podendo assim decidir se continua investindo em campanhas pagas, se muda sua estratégia, diminui os gastos...;
2. A média de tempo gasto do visitante no portal;
3. A principal fonte de acesso do portal;
4. Qual o Sitema Operacional mais usado para acesar o portal e com isso decidir se vale a pena investir em uma nova APP;
5. Qual o dispositivo mais usado para acessar o portal e com isso fazer melhorias ou correções no seu web site;
6. Total de faturamento por dia.

Além de todas essas informações foi adicionado o gráfico de KPI que mostra como está a interação do visitante no potal do e-commerce. Neste caso a meta não foi alcançada o que pode gerar um grande impacto na empresa. Reuniões com outros setores da empresa será necessário para tomadas de decisão. Esse gráfico também explica como a empresa conseguiu atingir a meta e como não conseguiu se manter.

As informações estão sendo apresentadas de maneira global. O gráfico é interativo podendo assim filtrar as informações de acordo com cada necessidade.
